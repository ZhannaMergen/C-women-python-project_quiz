```python
QUESTIONS = [
    ("1. J’adore le sport. \nMoi aussi. Je fais ___ rugby tous les dimanches.", "du"),
    ("2. Voilà le menu. Tu veux du poisson ?\nNon, je n’aime pas ___ poisson.", "le"),
    ("3. _______ est ta nationalité ?", "quelle"),
    ("4. Vous voulez un apéritif ?\nNon merci. Je ne bois pas ___ alcool.", "d`"),
    ("5. Je suis fatigué de l`attendre. Il est toujours ____. ", "en retard"),
    ("6. Pardon Madame, vous avez ___ s’il vous plaît ?\nOui, il est 14h30.", "l`heure"),
    ("7. J’aime beaucoup ___ fruits !", "les"),
    ("8. Je viens ______ Allemagne.", "d`"),
    ("9. Je suis bavard, j’aime beaucoup ___.", "parler"),
    ("10. Moi, ce que je préfère, c’est voyager ___ avion.", "en"),
    ("11.Comment ? Tu ne ___ souviens pas du nom de ton voisin ?", "te"),
    ("12. Ça fait longtemps que vous étudiez le français ?\nOui, je l’étudie ___ cinq ans.", "depuis"),
    ("13. Tu connais un bon restaurant de poissons ?\n ___ qui est sur la place est excellent. C’est le plus connu de la région.", "celui"),
    ("14. Qu’est-ce que ton bébé prend au goûter ? \nUne compote ___ pommes.", "de"),
    ("15. Madame, je peux vous ___ une question ?\nBien-sûr. Les professeurs sont là pour ça.", "poser"),
    ("16. On peut vraiment compter ___ lui. C’est un homme sérieux.", "sur"),
    ("17. Tu as déjà ___ des romans policiers ?", "lu"),
    ("18. J’ai étudié tous les cours ___ tu m’as prêtés. Ils m’ont beaucoup servi.", "que"),
    ("19. Oh, regarde le joli timbre.\nAh, oui ! Tiens ! La lettre vient ___ Portugal.", "du"),
    ("20. Vous parlez l’anglais ?\nOui, un peu, mais mon collègue parle ___ l’anglais que moi.","mieux"),
    ("21. On est quel jour aujourd’hui ? \nAujourd’hui ? On est ___ 28 janvier.", "le"),
    ("22. Est-ce que tu crois que je suis jalouse ___elle ?", "d`"),
    ("23. Est-ce que vous connaissez l’objectif de cette association ?\nOui, elle lutte ___ le racisme.", "contre"),
    ("24. Tu peux lui demander conseil. Elle est tres douée ___ informatique.", "en"),
    ("25. Je suis d’accord avec toi. Je pense que tu ___ absolument raison.", "as"),
    ("26. Il faut que tu ___ voir un dentiste de toute urgence.", "ailles"),
    ("27. Jeudi, je ne peux pas. ___ vendredi, c’est parfait, je suis disponible.", "par contre"),
    ("28. Je suis sûr que je vais te battre aux échecs. Je suis ___ que toi !", "meilleur"),
    ("29. Bonjour madame. Vous désirez ? \n  ____-je voir le responsable du magasin, s’il vous plaît ?", "puis"),
    ("30. Tu as une nouvelle voiture ?\nTu me ___ montres ?", "la"),

]

number = 0
for x in range(10):
    print()
    answer = input("Are you ready to pass this quiz : yes/no").lower()
    if answer =='no':
      print()
      print("You can pass it later! Have a nice day!")
      print()
      continue
    else:
      score = 0
    print()
    Lname = input("What is your last name?")
    Fname = input ("What is your first name?")
    Phone = input ("What is your phone number?")

    print() 
    print("Fill in the blanks to complete sentences")
    print()
    for question, correct_answer in QUESTIONS:
     answer = input(f"{question}").lower()
     if answer == correct_answer:
       print("Correct!")
       score = score+1
     else:
      print(f"The answer is {correct_answer!r}, not {answer!r}")
    print ("Your score is", score, "/30")

    print("You answered", str(score/30*100), "%.")
    
    if score < 5:
      print ("You have to start from scratch.")
      level = "A0"
    elif score >= 5 and score <12:
      print("You have A1 level or beginner level. You  can understand familiar names, words, and basic phrases–for example, on posters, catalogs, signs, or graphics.\nYou can write short, simple sentences, like what I might write on a postcard. You could fill in your name, nationality, and address on a hotel registration form.")
      level = "A1"
    elif score >=12 and score < 20:
      print ("You have A2 level or elementary level. You  can read short, simple texts about everyday information such as menus, printed instructions, transportation time tables, or informational text messages. \nYou can write simple notes and messages, such as text messages, personal thank you letters, or grocery lists.")
      level = "A2"
    elif score>=20:
      print("You have B1 or intermediate level.You  can understand texts that consist of every-day languages, such as descriptions of recent events in a newspaper article, memos about your area of work or study, \nor texts from friends recounting their day. You can write a text on a topic that’s of personal interest, such as describing your experiences or impressions of something that happened to you.")
      level = "B1"
    number = number+1
    NameScore=str(number) + " " + Lname + "  " + Fname + " " + str(score) + " " + level + " " + str(Phone)
    fname = 'Student_List.txt'  #append to an existing file
    with open(fname, 'a') as file_object: #a means append
      file_object.write(NameScore + "\n")
  
```
